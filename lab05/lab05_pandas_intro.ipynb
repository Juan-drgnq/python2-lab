## Ćwiczenie 1: Series — tworzenie i operacje

### Krok 1

```python
oceny = pd.Series([4.5, 3.0, 5.0, 3.5, 4.0])
print(oceny)
```

```python
pensje = pd.Series({'Anna': 5500, 'Jan': 7200, 'Ewa': 4800, 'Marek': 9100, 'Kasia': 6300})
print(pensje)
```

```python
#1
print(pensje.idxmax())
#2
print(f'Średnia pensja wynosi: {pensje.mean()}')
#3
print(f'Osoby zarabiające powyżej średniej to:\n{pensje[pensje > pensje.mean()]}')
#4
print(f'Liczba osób zarabiająca < 6000 to: {(pensje[pensje < 6000]).count()}')
```

### Krok 2

```python
#1
pensje_podwyzka = pensje * 1.10
print(f'Podwyzka:\n{pensje_podwyzka}')
#2
pensje_premia = pensje + 500
print(f'Premia:\n{pensje_premia}')
#3
nowa_pensja = (pensje * 1.10) + 500
print(f'Nowa:\n{nowa_pensja}')
#4
print(f'Sort:\n{nowa_pensja.sort_values(ascending=False)}')
```

## Ćwiczenie 4: Pytania biznesowe

### Zad.4a:

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
# Dodaj kolumnę z procentem napiwku
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)
```

```python
# 1. Oblicz średni rachunek (total_bill)
print(f'Sredni rachunek: {tips.total_bill.mean()}')
# 2. Oblicz średni napiwek w procentach (tip_pct)
print(f'Sredni napiwek w proc: {tips.tip_pct.mean()}\n')
# 3. Wyświetl który dzień ma najwięcej rachunków (value_counts na 'day') (Sat)
print(f'{tips.day.value_counts()}\n')
# 4. Policz ile rachunków jest od palaczy (value_counts na 'smoker') (93)
print(f'{tips.smoker.value_counts()}\n')
# 5. Wyświetl describe() tylko dla kolumny 'tip' — jaka jest mediana napiwku? (2.9)
print(f'{tips.describe()}')
```

```python
# 6. Jaki procent rachunków pochodzi z obiadu (Dinner)? (72%)
print(f'{tips.time.value_counts(normalize=True)}\n')
# Podpowiedź: value_counts(normalize=True) na kolumnie 'time'
# 7. Ile unikalnych rozmiarów grup jest w danych? (nunique na 'size')
print(f'Unikalnych rozmiarow grup: {tips['size'].nunique()}')
```
