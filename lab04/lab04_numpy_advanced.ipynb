## Ćwiczenie 1: Broadcasting — zastosowania biznesowe

```python
ceny = np.array([[29.99, 31.99, 28.99], # Produkt A
[49.99, 52.99, 47.99], # Produkt B
[14.99, 15.99, 13.99], # Produkt C
[99.99, 104.99, 94.99]]) # Produkt D
sklepy = ['Centrum', 'Galeria', 'Online']
produkty = ['A', 'B', 'C', 'D']
print(f"Kształt macierzy: {ceny.shape}")
```

```python
#Zadanie 1a
ceny_inflacja = ceny * 1.05
print(ceny_inflacja)
```

```python
#Zadanie 1b
rabaty = np.array([0.10, 0.05, 0.15])
ceny_rabat = ceny * (1-rabaty)
print(ceny_rabat)
```

```python
#Zadanie 1c
marza = np.array([[0.30], [0.40], [0.25], [0.50]])
ceny_marza = ceny * (1 + marza)
print(ceny_marza)
```

```python
#Zadanie 1d
najtansze = ceny.argmin(axis=1)
y = 1
for x in najtansze:
    print(f'Najtanszy sklep dla Produktu-{y} to {sklepy[x]}')
    y += 1
```

## Ćwiczenie 2: Reshape i łączenie danych

```python
#Zadanie 2a
# Dane miesięczne z jednego roku (flat)
dane_flat = np.array([120, 135, 98, 142, 167, 189, 201, 178, 156, 143, 198, 221])
```

```python
kwartaly = dane_flat.reshape(4, 3)
print(f"Kwartały:\n{kwartaly}")
print(f"Suma per kwartał: {kwartaly.sum(axis=1)}")
print(f"Najlepszy kwartał: Q{kwartaly.sum(axis=1).argmax() + 1}")
```

```python
#Zadanie 2b — Stacking
# Dane z 3 oddziałów (osobne tablice)
warszawa = np.array([150, 160, 170, 180])
krakow = np.array([120, 130, 140, 150])
wroclaw = np.array([90, 100, 110, 120])
```

```python
# Połącz w macierz (3 oddziały × 4 kwartały)
wszystkie = np.vstack([warszawa, krakow, wroclaw])
print(f"Wszystkie:\n{wszystkie}")
print(f"Suma per oddział: {wszystkie.sum(axis=1)}")
print(f"Suma per kwartał: {wszystkie.sum(axis=0)}")
```

```python
#Zadanie 2c
sprzedaz_flat = np.array([100, 200, 150, 300, 250, 180, 120, 220, 170, 310, 260, 190])
#1 - Reshape na macierz 6×2
sprzedaz_macierz = sprzedaz_flat.reshape(6,2)
print(sprzedaz_macierz)
#2 - Oblicz średnią sprzedaż per produkt (axis=1)
srednie_ceny = sprzedaz_macierz.mean(axis=1)
print(srednie_ceny)
#3 - Oblicz zmianę między miesiącem 1 a 2 (kolumna 1 - kolumna 0)
print(sprzedaz_macierz[:, 1] - sprzedaz_macierz[:, 0])
```

```python
#Zadanie 2d
nowe = wszystkie.flatten()
nowe = np.concatenate([nowe, np.array([235])])
print(nowe)
```

## Ćwiczenie 3: Analiza danych finansowych

```python
#Dane
np.random.seed(2026)

# 8 pracowników, dane roczne
pracownicy = ['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek']
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.0, 4.7, 3.5, 4.3, 3.9])

# Sprzedaż miesięczna 4 produktów przez 6 miesięcy
produkty_nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Akcesorium']
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_produktow = np.array([3500, 1800, 2500, 150])
```

```python
#Zadanie 3a
#1 - Oblicz średnią, medianę i odchylenie standardowe pensji
srednia = pensje.mean()
mediana = np.median(pensje)
odchylenie = pensje.std()
print(f'Srednia: {srednia} \nMediana: {mediana} \nOdchylenie: {odchylenie}')
print('\n')

#2 - Wyświetl imiona osób zarabiających powyżej średniej
pracownicy_arr = np.array(pracownicy)
maska_pracownicy = pensje > pensje.mean()
print(pracownicy_arr[maska_pracownicy])
print('\n')

#3 - Wyświetl imiona osób zarabiających poniżej mediany
maska_mediana = pensje < np.median(pensje)
print(pracownicy_arr[maska_mediana])
print('\n')

#4 - Oblicz rozstęp (max - min) i rozstęp międzykwartylowy (IQR)
rozstep = pensje.max() - pensje.min()
iqr = np.percentile(pensje,75) - np.percentile(pensje,25)
print(f'Rozstęp: {rozstep}, Rozstep miedzykwartylowy: {iqr}')
```

```python
#Zadanie 3b
premie = np.where(oceny >= 4.5, 2000, np.where(oceny >= 4.0, 1000, 0))
wynagrodzenie = pensje + premie
print(wynagrodzenie)
print(pracownicy_arr[premie.argmax()])
```

```python
#Zadanie 3c
kolejnosc = np.argsort(pensje)[::-1]
for i, idx in enumerate(kolejnosc,1):
    print(f'Imie - {pracownicy_arr[idx]}, Pensja - {pensje[idx]}, Staz - {staz[idx]} lat')
```

```python
#Zadanie 3d
# Oblicz korelację: staż vs pensja
print(np.corrcoef(staz,pensje)[0,1])
# Oblicz korelację: staż vs ocena
print(np.corrcoef(staz,oceny)[0,1])
#Silniejszy jest zwiazek staz vs pensja
```

```python
#Zadanie 3e
ceny_kol = ceny_produktow.reshape(4, 1)
przychod = sprzedaz * ceny_kol
print(przychod.argmax(axis=1))
print(przychod.argmax(axis=0))
print(przychod[:, :3].sum())
print(przychod[:, 3:].sum())
```

## Ćwiczenie 4: Generowanie danych

```python
np.random.seed(42)
wiek = np.clip(np.random.normal(35, 10, 100).astype(int), 18, None)
wydatki = np.clip(np.random.normal(500, 200, 100), 0, None)
wizyty = np.random.poisson(5, 100)
```

```python
#1 - Statystyki opisowe każdej zmiennej (mean, median, std, min, max)
print(f"Wiek — średnia: {wiek.mean():.1f}, mediana: {np.median(wiek):.1f}, odchylenie: {wiek.std():.1f}, min: {wiek.min()} max: {wiek.max()}")
#2 - Korelacja wiek vs wydatki
print(f'Korelacja wiek vs wydatki: {np.corrcoef(wiek, wydatki)[0, 1]:.4f}')
```

```python
#3 - Korelacja wizyty vs wydatki
print(f'Korelacja wiek vs wydatki: {np.corrcoef(wizyty, wydatki)[0, 1]:.4f}')
```

```python
sr = wydatki.mean()
sd = wydatki.std()
segment = np.where(wydatki > sr + sd, 'WYSOKI', np.where(wydatki < sr - sd, 'NISKI','ŚREDNI'))
print(segment)
wartosci = np.unique(segment, return_counts=True)
print(wartosci)
```
